```python
#Task 3
#Line Chart
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('top-1000-trending-youtube-videos.csv')

# If 'published' just contains years like 2017, 2018, etc.
data['year'] = data['published'].astype(int)

# Verify years
print(data['year'].value_counts().sort_index())

# Plot line chart
yearly_counts = data['year'].value_counts().sort_index()
plt.plot(yearly_counts.index, yearly_counts.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Videos Published')
plt.title('Trending YouTube Videos Published Per Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Bar Chart
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('top-1000-trending-youtube-videos.csv')

# Clean numeric values
data['Video views'] = pd.to_numeric(data['Video views'].str.replace(',', ''), errors='coerce')

# Clean and standardize category names
data['Category'] = data['Category'].astype(str).str.strip().str.title()

# Drop rows with missing categories or video views
data = data.dropna(subset=['Category', 'Video views'])

# Group and calculate mean
grouped = data.groupby('Category')['Video views'].mean().sort_values(ascending=False)
print(grouped)

# Plot
plt.figure(figsize=(12, 6))
plt.bar(grouped.index, grouped.values, color='skyblue')
plt.xlabel('Category')
plt.ylabel('Average Video Views')
plt.title('Average Video Views per Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()



```

```python
import numpy as np

plt.figure(figsize=(10, 6))
plt.hist(np.log10(data['Video views'] + 1), bins=50, color='coral', edgecolor='black')
plt.xlabel('Log10(Video Views + 1)')
plt.ylabel('Frequency')
plt.title('Distribution of YouTube Video Views (Log Scale)')
plt.tight_layout()
plt.show()
```

```python
# Load data
data = pd.read_csv('top-1000-trending-youtube-videos.csv')

# Clean numeric columns
data['Video views'] = pd.to_numeric(data['Video views'].str.replace(',', ''), errors='coerce')
data['Likes'] = pd.to_numeric(data['Likes'].str.replace(',', ''), errors='coerce')

# Drop missing values
data = data.dropna(subset=['Video views', 'Likes'])

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Video views'], data['Likes'], alpha=0.5, color='purple')
plt.xlabel('Video Views')
plt.ylabel('Likes')
plt.xscale('log')
plt.yscale('log')
plt.title('Relationship Between Video Views and Likes')
plt.grid(True)
plt.tight_layout()
plt.show()
```
